```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-0.1, 3.1, 0.1)
y = np.exp(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# num2 = [np.exp(0.0), np.exp(0.1), np.exp(0.2), np.exp(0.3), np.exp(0.4)]
# print(num2)
x = np.arange(0.0,0.6,0.1)
y = np.exp(x)
print(y)
```

```python
plt.plot(x,y, label="exp")
plt.plot(x,x, linestyle='--', label="equ")
plt.xlabel("x")
plt.ylabel("y")
plt.title("exp & equ")
plt.legend()
plt.show()
```
